```python
# How many of each race are represented in this dataset? This should be a Pandas series with race names as the index labels
race_count = df['race'].value_counts().count()
race_count
```

```python
 # What is the average age of men?
average_age_men = df[df['sex'] == 'Male']['age'].mean()
average_age_men
```

```python
 # What is the percentage of people who have a Bachelor's degree?
percentage_bachelors = (df[df['education'] == 'Bachelors']['education'].shape[0] / len(df)) * 100
percentage_bachelors
```

```python
# What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K?
# What percentage of people without advanced education make more than 50K?

# with and without `Bachelors`, `Masters`, or `Doctorate`
higher_education = df[df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]
lower_education = df[~df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]

higher_education_above_50k = higher_education[higher_education['salary'] == '>50K']
lower_education_above_50k = lower_education[lower_education['salary'] == '>50K']

higher_education_rich = (len(higher_education_above_50k) / len(higher_education)) * 100
lower_education_rich = (len(lower_education_above_50k) / len(lower_education)) * 100

higher_education_rich
lower_education_rich
```

```python
# What is the minimum number of hours a person works per week (hours-per-week feature)?
min_work_hours = df['hours-per-week'].min()

min_work_hours
```

```python
df[df['hours-per-week'] == 1]
```

```python
 # What percentage of the people who work the minimum number of hours per week have a salary of >50K?
num_min_workers = df[df['hours-per-week'] == 1]
min_hour_rich = num_min_workers[num_min_workers['salary'] == '>50K']

rich_percentage = (len(min_hour_rich) / len(num_min_workers)) * 100

rich_percentage
```

```python
# What country has the highest percentage of people that earn >50K?
country_group = df.groupby('native-country')

total_by_country = country_group.size()

above_50k_by_country = country_group.apply(lambda x: (x['salary'] == '>50K').sum())

percentage_by_country = (above_50k_by_country / total_by_country) * 100

highest_earning_country = percentage_by_country.idxmax()
highest_earning_country_percentage = percentage_by_country.max()
```
